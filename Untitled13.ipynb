```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the weather dataset
url = "https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip"
df = pd.read_csv(url)

# Inspect the first few rows of the dataset
print(df.head())

# Display basic information about the dataset
df.info()

# Display summary statistics
print(df.describe())
```

```python
# Parse the date-time column
df['Date Time'] = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')

# Set the 'Date Time' column as the index
df.set_index('Date Time', inplace=True)

# Extract temperature data ('T (degC)' column)
temperature = df['T (degC)']

# Inspect the temperature column
print(temperature.head())
```

```python
# Plot the entire temperature time series
plt.figure(figsize=(10, 6))
temperature.plot(label='Temperature (°C)')
plt.title('Temperature Time Series')
plt.xlabel('Date Time')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Plot the temperature for the first 10 days
plt.figure(figsize=(10, 6))
temperature[:10*24].plot(label='Temperature (°C)')  # Assuming data is recorded hourly
plt.title('Temperature Time Series (First 10 Days)')
plt.xlabel('Date Time')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```
